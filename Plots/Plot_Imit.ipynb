```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_Acc = pd.read_csv('/content/Accuracy.csv')
data_Loss = pd.read_csv('/content/Loss.csv')
```

```python
# plt.plot(data['Train_Accuracy'], data['Step'], label='Train Accuracy', color='blue')
plt.plot(data_Acc['Step'], data_Acc['Validation_Accuracy'], label='Validation Accuracy')
plt.plot(data_Acc['Step'], data_Acc['Train_Accuracy'], label='Training Accuracy')
plt.xlabel('Training Accuracy')
plt.ylabel('Steps')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# plt.plot(data['Train_Accuracy'], data['Step'], label='Train Accuracy', color='blue')
plt.plot(data_Loss['Step'], data_Loss['Validation_Loss'], label='Validation Loss')
plt.plot(data_Loss['Step'], data_Loss['Training_Loss'], label='Training Loss')
plt.xlabel('Training Loss')
plt.ylabel('Steps')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.show()
```
